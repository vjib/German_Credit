```python
#For Tree model

client_input=[]
client_output=[]

for index,row in df_credit.iterrows():
    
    temp_input=[0]*9
    
    age=row['Age']    
    gender=row['Sex']
    
    if gender=='male':
        gender=1
    else:
        gender=2
    
    job=row['Job']
    housing=row['Housing']
    
    if housing=='free':
        housing=0
    elif housing=='own':
        housing=1
    elif housing=='rent':
        housing=2
    else:
        housing=3
        
    saving=row['Saving accounts']
    
    if saving=='little':
        saving=0
    elif saving=='moderate':
        saving=1
    elif saving=='quite rich':
        saving=2
    elif saving=='rich':
        saving=3
    else:
        saving=4
    
    checking=row['Checking account']
    
    if checking=='little':
        checking=0
    elif checking=='moderate':
        checking=1
    elif checking=='rich':
        checking=2
    else:
        checking=3

    credit=row['Credit amount']
    duration=row['Duration']
    purpose=row['Purpose']
    
    if (purpose=="business"):
        purpose=0
    elif (purpose=="car"):
        purpose=1
    elif (purpose=="domestic appliances"):
        purpose=2
    elif (purpose=="education"):
        purpose=3
    elif (purpose=="furniture/equipment"):
        purpose=4
    elif (purpose=="radio/TV"):
        purpose=5
    elif (purpose=="repairs"):
        purpose=6
    elif (purpose=="vacation/others"):
        purpose=7
    else:
        purpose=8
    
    temp_input[0]=age
    temp_input[1]=gender
    temp_input[2]=job
    temp_input[3]=housing
    temp_input[4]=saving
    temp_input[5]=checking
    temp_input[6]=credit
    temp_input[7]=duration
    temp_input[8]=purpose
 
    risk=row['Risk']
    
    client_input.append(temp_input)
    client_output.append([risk])
```

```python
client_input_np=np.array(client_input)
client_output_np=np.array(client_output)
```

```python
from sklearn.model_selection import train_test_split
client_input_train,client_input_test, client_output_train,client_output_test = train_test_split(client_input_np, client_output_np, test_size=0.2)
```

```python
#For NN model

client_input=[]
client_output=[]

for index,row in df_credit.iterrows():
    
    temp_input=[0]*22
    
    age=row['Age']    
    gender=row['Sex']
    
    if gender=='male':
        temp_input[0]=1
    
    temp_input[1]=(age-19)/(75-19)
    
    job=row['Job']
    
    temp_input[2+job]=1
    
    housing=row['Housing']
    
    if housing=='free':
        temp_input[6]=1
    elif housing=='own':
        temp_input[7]=1
    elif housing=='rent':
        temp_input[8]=1
    else:
        temp_input[9]=1
        
    saving=row['Saving accounts']
    
    if saving=='little':
        temp_input[10]=0.25
    elif saving=='moderate':
        temp_input[10]=0.5
    elif saving=='quite rich':
        temp_input[10]=0.75
    elif saving=='rich':
        temp_input[10]=1.0
    else:
        temp_input[10]=0
    
    checking=row['Checking account']
    
    if checking=='little':
        temp_input[11]=0.33
    elif checking=='moderate':
        temp_input[11]=0.67
    elif checking=='rich':
        temp_input[11]=1.0
    else:
        temp_input[11]=0

    credit=row['Credit amount']
    duration=row['Duration']
    
    temp_input[12]=(credit-250)/(18424-250)
    temp_input[13]=(duration-4)/(72-4)
    
    purpose=row['Purpose']
    

    if (purpose=="business"):
        temp_input[14]=1
    elif (purpose=="car"):
        temp_input[15]=1
    elif (purpose=="domestic appliances"):
        temp_input[16]=1
    elif (purpose=="education"):
        temp_input[17]=1
    elif (purpose=="furniture/equipment"):
        temp_input[18]=1
    elif (purpose=="radio/TV"):
        temp_input[19]=1
    elif (purpose=="repairs"):
        temp_input[20]=1
    elif (purpose=="vacation/others"):
        temp_input[21]=1
   
    client_input.append(temp_input)

    risk=row['Risk']
    
    if risk=='good':
        client_output.append([0])
    else:
        client_output.append([1])
```

```python
client_input_np=np.array(client_input)
client_output_np=np.array(client_output)
```

```python
from sklearn.model_selection import train_test_split
client_input_train,client_input_test, client_output_train,client_output_test = train_test_split(client_input_np, client_output_np, test_size=0.2)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
import tensorflow as tf

config = tf.ConfigProto( device_count = {'GPU': 1 , 'CPU': 56} ) 
sess = tf.Session(config=config) 
keras.backend.set_session(sess)
```

```python
# Neural network
model = Sequential()
model.add(Dense(100, input_dim=len(client_input_train[0]), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(100, activation='relu'))
model.add(Dense(len(client_output_train[0]), activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```
